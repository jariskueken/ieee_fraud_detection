```python
from collections import Counter
from imblearn.over_sampling import SMOTE
import pandas as pd
```

```python
train_df = pd.read_csv('../data/train_data_prepared.csv', index_col=None)
```

```python
# create features and target lists for the dataset
features = list(train_df.columns)
print(features)
features.remove('isFraud')
target = 'isFraud'

# print the shape of the original array
print(train_df.shape)
```

```python
# read the target
y = train_df[target].to_numpy()

counter = Counter(y)
print(counter)
```

```python
print(features)
train_df.head()
```

```python
# oversample using SMOTE
feature_vals = train_df[features]
target_vals = train_df[target]
oversample = SMOTE(random_state=42, k_neighbors=5)
X_oversampled, y_oversampled = oversample.fit_resample(feature_vals, target_vals)

# create new dataframes
train_resampled = pd.DataFrame(data=X_oversampled, columns=features)
train_resampled_target = pd.DataFrame(data=y_oversampled)

train_resampled.head()
train_resampled = pd.concat([train_resampled, train_resampled_target], axis=1)

print(train_resampled.shape)
train_resampled.head()
```

```python
counter_resampled = Counter(train_resampled[target])
print(counter_resampled)
```

```python
# write the oversampled data to a csv
train_resampled.to_csv('../data/train_data_resampled.csv', index=False)
```
